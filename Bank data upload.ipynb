```python
for index, row in df.iterrows():
    if District.objects.filter(name=row['District']).exists():
        district = District.objects.get(name=row['District'])
#         print(index, 'th row under', district.name)
    if Bank.objects.filter(bank_name=row['Name of the bank']).exists():
        bank = Bank.objects.get(bank_name=row['Name of the bank'])
#         print('this row comes with ', bank.bank_name,)
    print('{} name and District {}'.format(bank.bank_name, district.name))
    bank_branch_obj, created = BankBranch.objects.get_or_create(branch_name=row['Branch name'],
                                                                ifsc_code=row['Ifsc code'],
                                                                state_id=23,
                                                                address=row['Address'],
                                                                micr_code=row['Micr code'],
                                                                district=district,
                                                                bank=bank,
                                                                created_by='RBI Data')
    if created:
        print(index,'New Branch is Created for {} branch under {}, {} district'.format(bank_branch_obj.branch_name, bank.bank_name, district.name))
    else:
        print(index,'Branch already exists! Branch: {}, bank: {}, district: {}'.format(bank_branch_obj.branch_name, bank.bank_name, district.name))
# print(bank_branch_obj)
         
```
